```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
m=250.0
c=1000.0
k=800.0

def f(t):
    return np.sin(2*np.pi*t)

def system (x,t):
    x1,x1_dot=x
    x1_ddot=(k*f(t)-c*x1_dot-k*x1)/m
    return[x1_dot,x1_ddot]

x0=[0.0,0.0]
t=np.linspace(0,5,500)
solution = odeint(system,x0,t)
x=solution[:,0]

plt.plot(t,x,label='displacement(x)')
plt.plot(t,f(t),label='external force (f(t))',linestyle='dashed')
plt.xlabel('time[s]')
plt.ylabel('displacement[m]')
plt.title('mass sprint damper system')
plt.legend
plt.grid(True)
plt.show()
```
